### K-Means Clustering
aka. Hill Climbing Algorithm
* Assign
    - a center for the cluster of points
* Optimize
    - by taking the minimum quadratic distance (aka. rubber bands) of individual points from the center
    
The Output depends on the intial cluster centers(centroid).
So, the same training set may yield different output for various cluster centers.

Red - Data points

Cyan - Ideal centroid local minima

Green - Real local minima

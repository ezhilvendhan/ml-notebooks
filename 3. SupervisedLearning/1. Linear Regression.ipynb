### Absolute Trick

Move the line $y=w_1x + w_2$ absolutely in small steps using $y=(w_1+p\alpha)x + (w_2+\alpha)$

$\alpha$ is the learning rate. It is +ve if the point is the above the line and -ve if below.
Also, `p` is -ve is the point is on the left quadrant

Need to have 2 tricks - one above the point and the other below

### Square Trick

No 2 tricks needed as in Absolute trick as we factor in the vertical distance as below

$y=(w_1+p\alpha(q-q\prime))x + (w_2+\alpha(q-q\prime))$

### Mean Absolute Error functions

$$
Error = \frac {1}{m} \sum_{i=1}^m \mid y-\widehat{y}\mid
$$

where, m is the number of records.

We take the absolute value like $\mid y - \widehat{y}\mid$ to ensure that negative values don't square out the errors. 
This is nothing but the absolute trick.

### Mean Squared Error

$$
Error = \frac {1}{2m} \sum_{i=1}^m (y-\widehat{y})^2
$$

The half doesn't matter much and is there only for convenience during derivates. 
This is nothing but the squared trick.

### Mean vs Total Squared (or Absolute) Error

Relation between Total squared(or absolute) error T and Mean squared(or absolute) error M is
$$M = mT$$

Since derivatives are linear functions, the gradient of T is also m times the gradient of M.

Choosing between the mean squared error and the total squared error really just amounts to picking a different learning rate.
